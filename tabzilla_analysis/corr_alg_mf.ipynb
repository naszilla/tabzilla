# Finding the  most correlated features for a particular algorithm

```python
def alg_corrs(alg):
    
    alg_df = joined_df.loc[joined_df['alg_name'] == alg]
    corr_columns = [col_name for col_name in metafeatures_df if col_name.startswith("f__")]
    return (alg_df[corr_columns].corrwith(alg_df[f"{metric}__test"])).abs().sort_values(ascending=False)
    
```

```python
def feature_corr(algos):

    df = pd.DataFrame(columns = [ 'Feature 1', 'Feature 2'])
    pd.set_option('max_colwidth', 400)
    for algo in algos:
        features = (alg_corrs(algo).index[:2]).tolist()
        scores = alg_corrs(algo).tolist()[:2]
        feature_scores = [f"{features[i]}, score:{scores[i]}" for i in range(2)]
        df.loc[algo] = feature_scores
    return df
```

```python
feature_corr(metadataset_df['alg_name'].unique())
```
